# Stability-AI/lm-evaluation-harness gptq model example

```python
%%bash
git clone -b jp-stable https://github.com/Stability-AI/lm-evaluation-harness.git
cd lm-evaluation-harness
pip install -e ".[ja]"
```

```python
%%bash
cd lm-evaluation-harness

# If you do not manually install gekko, the autoGPTQ installation may fail.
pip install gekko
pip install -e ".[auto-gptq]"
```

```python
%%bash
cd lm-evaluation-harness
PROJECT_DIR="dahara1"
MODEL_NAME="weblab-10b-instruction-sft-GPTQ"
PRETRAINED="${PROJECT_DIR}/${MODEL_NAME}"
MODEL_ARGS="pretrained=${PRETRAINED},quantized=gptq_model-4bit-128g.safetensors,gptq_use_triton=True"

TASK="jsquad-1.1-0.3"
NUM_FEWSHOT="2"
OUTPUT_PATH="result.json"
python main.py \
	    --model hf-causal-experimental \
	    --model_args $MODEL_ARGS \
	    --tasks $TASK \
            --num_fewshot $NUM_FEWSHOT \
            --device "cuda" \
            --output_path $OUTPUT_PATH
```

```python
%%bash
cd lm-evaluation-harness
cat result.json
```
